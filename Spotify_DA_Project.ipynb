```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks=pd.read_csv('tracks.csv')
df_tracks.head()
```

```python
#null values 
pd.isnull(df_tracks).sum()
```

```python
df_tracks.describe().transpose()
```

```python
#top 10 least popular songs
df_tracks.sort_values('popularity',ascending=True).head(10)
```

```python
#most popular songs with popularity greater than 90
df_tracks.query('popularity>90').sort_values('popularity',ascending=False).head(10)
```

```python
#setting release date as index of the table
df_tracks.set_index('release_date',inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head(3)
```

```python
#checking the artist who is at 18th location
df_tracks[['artists']].iloc[18]
```

```python
#changing duration from miliseconds to seconds
df_tracks['duration']=df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms',axis=1,inplace=True)
df_tracks.head(3)
```

```python
corr_df=df_tracks.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='inferno',linewidths=1,linecolor='black')
heatmap.set_title("Corrlation heatmap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
#using 0.4% of data for regression plot between loudness & energy
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness',x='energy',color='c').set(title='Loudness vs Energy')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='popularity',x='acousticness',color='g').set(title='Popularity vs Acousticness')
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
total_dr=df_tracks.duration
fig_dim=(18,7)
fig,ax=plt.subplots(figsize=fig_dim)
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title="Years vs Duration")
plt.xticks(rotation=90)
```

```python
total_dr=df_tracks.duration
fig_dim=(18,7)
fig,ax=plt.subplots(figsize=fig_dim)
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title="Years vs Duration")
plt.xticks(rotation=60)
```

```python
df_genre=pd.read_csv('SpotifyFeatures.csv')
df_genre.head()
```

```python
plt.title("Duration of Songs in Different Genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y='genre',x='duration_ms',data=df_genre)
plt.xlabel("Duration in milliseconds")
plt.ylabel("Genre")
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous=df_genre.sort_values("popularity",ascending=False).head(10)
sns.barplot(y='genre',x='popularity',data=famous).set(title="Top 5 popular genres")
```
